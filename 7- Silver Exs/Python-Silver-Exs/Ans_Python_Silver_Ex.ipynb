## 3. Assignment Operators

```python
# Input the initial value
initial_value = float(input("Enter the initial value: "))

# Apply different assignment operators
initial_value += 5  # Add 5 to the initial value
print("After += 5, the value is:", initial_value)

initial_value -= 3  # Subtract 3 from the current value
print("After -= 3, the value is:", initial_value)

initial_value *= 2  # Multiply the current value by 2
print("After *= 2, the value is:", initial_value)

initial_value /= 4  # Divide the current value by 4
print("After /= 4, the value is:", initial_value)

initial_value **= 3  # Raise the current value to the power of 3
print("After **= 3, the value is:", initial_value)

initial_value //= 2  # Perform floor division of the current value by 2
print("After //= 2, the value is:", initial_value)

initial_value %= 3  # Take the modulus of the current value by 3
print("After %= 3, the value is:", initial_value)
```

## 21. Creating and Modifying Sets

```python
# set with tuple product information
product_set = {("Laptop", 5), ("vacuum", 2)}

# Create a new empty set to store updated product information
new_product_set = set()

for product in product_set:
    name, quantity = product
    
    # Check if the product name is "Laptop"
    if name == "Laptop":
        quantity += 5
        product = name, quantity
        
    # Add the product tuple to the new set
    new_product_set.add(product)

# Print the updated product set
print(new_product_set)
```

## 33. Factorial

```python
def calculate_factorial(number):
    factorial = 1
    for i in range(1, number + 1):
        factorial *= i

    return factorial


# Print the result
print("Factorial:", calculate_factorial(5))
```

## 34. Counting Vowels

```python
def count_vowels(input_string):
    vowel_count = 0
    for char in input_string.lower():
        if char in 'aeiou':
            vowel_count += 1

    return vowel_count


input_string = "Hello World"
print("Number of vowels:", count_vowels(input_string))
```

## 35. Fibonacci Sequence

```python
def print_fibonacci_sequence(number):
    first = 0
    second = 1

    if number >= first:
        print(first)
    if number >= second:
        print(second)

    while (first + second) <= number:
        next_number = first + second
        print(next_number)
        
        first = second
        second = next_number


print("Fibonacci sequence:")
print_fibonacci_sequence(15)
```

## 36. Factorial

```python
def calculate_factorial(number):
    if number == 0:
        return 1

    return number * calculate_factorial(number - 1)

# Print the result
print("Factorial:", calculate_factorial(5))
```

## 37. Fibonacci Sequence

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1

    return fibonacci(n-1) + fibonacci(n-2)

# Print the result
print("Fibonacci sequence:", fibonacci(6))
```

## 39. File Reading

```python
def read_file(file_name):
    try:
        with open(file_name, 'r') as file:
            contents = file.read()
            print("File Contents:\n", contents)
    except FileNotFoundError:
        print("Error: File not found.")

file_name = input("Enter the file name: ")
read_file(file_name)
```

## 40. Input Validation

```python
def get_positive_integer():
    while True:
        try:
            number = int(input("Enter a positive integer: "))
            if number <= 0:
                raise ValueError
            return number
        except ValueError:
            print("Error: Invalid input. Please enter a positive integer.")


print("Valid positive integer entered:", get_positive_integer())
```
