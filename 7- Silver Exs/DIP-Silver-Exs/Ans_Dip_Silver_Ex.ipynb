## Question 1

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# a) Load a color image
image = cv2.imread('media/a.jpg')

# b) Split the image into its individual color channels
b_channel, g_channel, r_channel = cv2.split(image)

# Create images to display each channel in color
zeros = np.zeros_like(b_channel)

# Blue channel image (Blue channel with Red and Green channels set to 0)
blue_image = cv2.merge([b_channel, zeros, zeros])

# Green channel image (Green channel with Blue and Red channels set to 0)
green_image = cv2.merge([zeros, g_channel, zeros])

# Red channel image (Red channel with Blue and Green channels set to 0)
red_image = cv2.merge([zeros, zeros, r_channel])

# c) Display each color channel separately using its color
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Display Blue channel
axes[0].imshow(cv2.cvtColor(blue_image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Blue Channel')
axes[0].axis('off')

# Display Green channel
axes[1].imshow(cv2.cvtColor(green_image, cv2.COLOR_BGR2RGB))
axes[1].set_title('Green Channel')
axes[1].axis('off')

# Display Red channel
axes[2].imshow(cv2.cvtColor(red_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('Red Channel')
axes[2].axis('off')

plt.show()

# d) Merge the color channels back together inversely (BGR to RGB)
inversely_merged_image = cv2.merge([r_channel, g_channel, b_channel])

# Convert the inversely merged image from BGR to RGB for displaying with matplotlib
inversely_merged_image_rgb = cv2.cvtColor(inversely_merged_image, cv2.COLOR_BGR2RGB)

# Display the inversely merged image
plt.imshow(inversely_merged_image_rgb)
plt.title('Inversely Merged Image (RGB)')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# a) Load an image
image = cv2.imread('media/a.jpg')

# b) Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# c) Apply contrast-limited adaptive histogram equalization (CLAHE) to the grayscale
#    image using the createCLAHE() and apply() functions with customized parameters.
# Create a CLAHE object with customized parameters
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_image = clahe.apply(gray_image)

# d) Display the original and enhanced images side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Display the original grayscale image
axes[0].imshow(gray_image, cmap='gray')
axes[0].set_title('Original Grayscale Image')
axes[0].axis('off')

# Display the CLAHE enhanced image
axes[1].imshow(clahe_image, cmap='gray')
axes[1].set_title('CLAHE Enhanced Image')
axes[1].axis('off')

# Show the plot
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('media/d.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Find edges using Canny edge detection
edges = cv2.Canny(gray_image, 100, 200)

# Find lines using Hough Line Transform
lines = cv2.HoughLinesP(edges, 10, np.pi / 180, 150)

# Draw lines on the original image
lines_image = np.copy(image)
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(lines_image, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Display the original image with lines
# Create figure and axes
fig, axes = plt.subplots(1, 2, figsize=(20, 10))

# Display the original image
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')

# Display the lines
axes[1].imshow(cv2.cvtColor(lines_image, cv2.COLOR_BGR2RGB))
axes[1].set_title('Lines')
axes[1].axis('off')

# Show the plot
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('media/f.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# # Apply thresholding to obtain a binary image
# _, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Detect edges using Canny edge detection
edges = cv2.Canny(gray_image, 100, 255)  # Adjust the thresholds here

# Find contours in the binary image
contours, _ = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Create a blank image to draw the contours
contours_image_black = np.zeros_like(image)
contours_image = image.copy()

# Draw contours on the blank image
cv2.drawContours(contours_image_black, contours, -1, (0, 255, 0), 3)
cv2.drawContours(contours_image, contours, -1, (0, 255, 0), 3)

# Show the original image and the image with contours
# Create figure and axes
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

# Display the original image
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')

# Display the contours with black background
axes[1].imshow(contours_image_black)
axes[1].set_title('Contours with black background')
axes[1].axis('off')

# Display the contours
axes[2].imshow(cv2.cvtColor(contours_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('Contours')
axes[2].axis('off')

# Show the plot
plt.show()
```

## Question 13

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the video
video = cv2.VideoCapture('media/e.mp4')

# Set the frame number for the 18th and 19th frames
frame_numbers = [17, 18]

# Read and process the selected frames
frames = []
for frame_number in frame_numbers:
    # Set the frame number
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    
    # Read the frame
    ret, frame = video.read()
    
    if ret:
        # Convert the frame to grayscale
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        # Denoise the frame using a gaussian filter
        denoised_frame = cv2.GaussianBlur(gray_frame, (5, 5), 0)
        
        # Perform edge detection using the Canny edge detection method
        edges = cv2.Canny(denoised_frame, 50, 150)
        
        # Store the original frame and the edges in a tuple
        frames.append((frame, edges))
        
# Release the video object
video.release()

# Show the original frames with edges
# Create figure and axes
fig, axes = plt.subplots(len(frames), 2, figsize=(20, 5*len(frames)))

# Display the frames with edges
for i, (frame, edges) in enumerate(frames):
    axes[i, 0].imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    axes[i, 0].set_title(f'Frame {frame_numbers[i]}')
    axes[i, 0].axis('off')
    
    axes[i, 1].imshow(edges, cmap='gray')
    axes[i, 1].set_title('Edges')
    axes[i, 1].axis('off')
    
plt.tight_layout()
plt.show()
```

## Question 14

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img1 = cv2.imread('media/IMG1.jpg', cv2.IMREAD_ANYCOLOR)
img2 = cv2.imread('media/IMG2.jpg', cv2.IMREAD_ANYCOLOR)

img1 = cv2.resize(img1, (int(img1.shape[1]*0.4), int(img1.shape[0]*0.4)))
img2 = cv2.resize(img2, (int(img2.shape[1]*0.4), int(img2.shape[0]*0.4)))
```

```python
# SIFT
sift = cv2.SIFT_create()

# color mode
keypoint1, descriptor1 = sift.detectAndCompute(img1, mask=None)
keypoint2, descriptor2 = sift.detectAndCompute(img2, mask=None)

img1_kp = cv2.drawKeypoints(img1, keypoint1, 0, (255, 0, 0), flags= cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img2_kp = cv2.drawKeypoints(img2, keypoint2, 0, (255, 0, 0), flags= cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Brute force
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = bf.match(descriptor1, descriptor2)

matches = sorted(matches, key=lambda x:x.distance)

result = cv2.drawMatches(img1, keypoint1, img2, keypoint2, matches[:50], None)

_ = cv2.imwrite('media/sift_color_mode.jpg', result)

result = cv2.resize(result, (int(result.shape[1]*0.4), int(result.shape[0]*0.4)))
cv2.imshow('sift_color_mode', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Gray mode

# convert to gray
gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# SIFT
sift = cv2.SIFT_create()


keypoint1, descriptor1 = sift.detectAndCompute(gray_img1, mask=None)
keypoint2, descriptor2 = sift.detectAndCompute(gray_img2, mask=None)

img1_kp = cv2.drawKeypoints(gray_img1, keypoint1, 0, (255, 0, 0), flags= cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
img2_kp = cv2.drawKeypoints(gray_img2, keypoint2, 0, (255, 0, 0), flags= cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Brute force
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
matches = bf.match(descriptor1, descriptor2)

matches = sorted(matches, key=lambda x:x.distance)

result = cv2.drawMatches(img1, keypoint1, img2, keypoint2, matches[:50], None)
_ = cv2.imwrite('media/sift_gray_mode.jpg', result)

result = cv2.resize(result, (int(result.shape[1]*0.4), int(result.shape[0]*0.4)))
cv2.imshow('sift_color_mode', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# compare
img1 = cv2.imread('media/sift_color_mode.jpg', cv2.IMREAD_ANYCOLOR)
img2 = cv2.imread('media/sift_gray_mode.jpg', cv2.IMREAD_ANYCOLOR)

# Show the original image and the image with contours
# Create figure and axes
fig, axes = plt.subplots(1, 2, figsize=(20, 10))

# Display the original image
axes[0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
axes[0].set_title('Sift with color mode')
axes[0].axis('off')

# Display the contours with black background
axes[1].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
axes[1].set_title('Sift with gray mode')
axes[1].axis('off')

# Show the plot
plt.show()
```

```python
# compare reason:
# The gray level mode is better than the color level mode, although there are no major changes in the images used for comparison.
# In gray mode, light changes and brightness are better managed and we will see a better result. Also, when we use gray mode
# instead of color, we have only one channel instead of 3 color channels, which makes the SIFT algorithm less complicated and also
# in gray images. Compared to color images, it has less noise and it is better displayed in gray mode to find keypoints based on 
# neighboring intensities, including finding edges.
```
