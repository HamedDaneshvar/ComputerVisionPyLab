```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('Chambord-3.jpg', -1)
img2 = cv2.imread('Chambord-1.jpg', -1)
img3 = cv2.imread('Chambord-2.jpg', -1)
images = [img1, img2, img3]
for i in range(len(images)):
    images[i] = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(30, 20))
for i in range(len(images)):
    plt.subplot(1, 3, i+1)
    plt.axis('off')
    plt.imshow(images[i])
```

## Stich images together

```python
sticher = cv2.Stitcher_create()
status, result = sticher.stitch(images)
```

```python
if status == 0:
    plt.figure(figsize=(30, 20))
    plt.imshow(result)
```

```python
## save the result
result = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)
_ = cv2.imwrite('stiched_image.jpg', result)
```
