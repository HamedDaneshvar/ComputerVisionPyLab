```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### question 1

```python
# yes the original image need to enhance because image is dark
org_img = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)
org_img_hist = cv2.calcHist([org_img], [0], None, [256], (0, 256))
plt.figure(figsize=(4, 2), dpi=100)
plt.plot(org_img_hist)
plt.show()
# this plot showing us most pixel of image is dark and in range 0 to 5 (it's guess)
```

### question 2

```python
def show_img_plt(img, c_map ='gray'):
    plt.figure(dpi=100)
    plt.imshow(img, cmap=c_map)

def display_hist_plt(img, bins=256, range=(0, 256)):
    plt.figure(figsize=(4, 2), dpi=100)
    plt.hist(img.flat, bins=bins, range=range)
    plt.show()
```

#### Contrast Stretching method

```python
min_new = 50
max_new = 255

min_old = np.min(org_img)
max_old = np.max(org_img)

new_img = np.uint8( (((org_img-min_old)/(max_old-min_old))*205) + 50)
show_img_plt(new_img)

# Even using a minimum pixel intensity of 50 for the new image, the new image did not change
display_hist_plt(new_img)
```

#### Histogram Equalization method

```python
hist_equa_img = cv2.equalizeHist(org_img)
show_img_plt(hist_equa_img)

# use this method, The image has good contrast
display_hist_plt(hist_equa_img)
```

#### Gamma method

```python
# as gamma 0.1
gamma = 0.1
gamma_img =np.uint8(((org_img/255)**gamma)*255)
show_img_plt(gamma_img)

# use this method, image is better but we need increase gamma value to achieve better image. this image is very brighness
display_hist_plt(gamma_img)
```
