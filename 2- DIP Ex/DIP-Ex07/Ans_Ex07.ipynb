## Exercise 1

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_img_plt(img, c_map ='gray'):
    plt.figure(dpi=100)
    plt.imshow(img, cmap=c_map)

def display_hist_plt(img, bins=256, range=(0, 256)):
    plt.figure(figsize=(4, 2), dpi=100)
    plt.hist(img.flat, bins=bins, range=range)
    plt.show()
```

```python
org_img = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)
```

## Original Image with histogram

```python
show_img_plt(org_img)
display_hist_plt(org_img)
```

## CLAHE method with final histogram

### limit 100 (Not Answer)

```python
clahe_pattern = cv2.createCLAHE(clipLimit=100)
clahe_img = clahe_pattern.apply(org_img)
show_img_plt(clahe_img)
display_hist_plt(clahe_img)
```
